```python
 
import matplotlib.pyplot as plt
import numpy as np
 
 
def Mexican(X, R1, R2, C1, C2, epochs):
    W = np.zeros((len(X), len(X)))
    X_imput = X.copy()
    for i in range(len(X)):
        if i < R1+1:
            np.fill_diagonal(W[:, i:], C1)
            np.fill_diagonal(W[i:, :], C1)
        elif i < R2+1:
            np.fill_diagonal(W[:, i:], C2)
            np.fill_diagonal(W[i:, :], C2)
    plt.figure(figsize=(15, 10))
    plt.plot(X, label="input")
    print("input: ", X,"\n")
    for epoch in range(epochs):
        X = np.matmul(X, W)
        for i in range(len(X)):
            if X[i] > 2:
                X[i] = 2
            elif X[i] < 0:
                X[i] = 0
        print("X_",epoch+1,": ", X)
        plt.plot(X, label="epoch "+str(epoch+1))
    out=[]
    for i in range(len(X)):
        if i in X.nonzero()[0]:
            out.append(X_imput[i])
        else:
            out.append(0) 
    print("Output: ", out,"\n")
    plt.legend()
    plt.show()
 

X = np.array([0.32, 0.33, 0.28, 0.47, 0.66, 0.80, 0.4, 0.33, 0.1, 0.26])
Mexican(X=X, R1=0, R2=len(X), C1=0.6, C2=-0.01, epochs=10)
Mexican(X=X, R1=1, R2=5, C1=0.6, C2=-0.4, epochs=10)
```
