```python
def sigmoid(x):
    return 1.0/(1.0 + np.exp(-x))

def sigmoid_prime(x):
    return sigmoid(x)*(1.0-sigmoid(x))

def tanh(x):
    return np.tanh(x)

def tanh_prime(x):
    return 1.0 - x**2
def sign(x):
    out=[]
    for i in range(len(x)):
        if x[i]>=0:out.append(1)
        else: out.append(-1)
    return np.array(out)
```
